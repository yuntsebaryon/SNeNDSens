```python
import ROOT
```

```python
f = ROOT.TFile.Open("/Users/yuntse/data/coherent/SNeNDSens/g4/NueArCC/nueArCC_sns_yDir_g4_0000.root")
```

```python
events = f.Get("EDepSimEvents")
```

```python
event = ROOT.TG4Event()
events.SetBranchAddress("Event",ROOT.AddressOf(event))
```

```python
events.GetEntry(15)
```

```python
print(f'event number {event.EventId}')
print(f'number of trajectories {event.Trajectories.size()}')
```

```python
for p in event.Trajectories:
    print(f'{p.GetTrackId()}, {p.GetParentId()}, {p.GetPDGCode()}, ({p.GetInitialMomentum().Px()}, {p.GetInitialMomentum().Py()}, {p.GetInitialMomentum().Pz()}, {p.GetInitialMomentum().E()})')
    for t in p.Points:
        print(f'p = ({t.GetMomentum().Px()}, {t.GetMomentum().Py()}, {t.GetMomentum().Pz()}), {t.GetProcess()}')
```
