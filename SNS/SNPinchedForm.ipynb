```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fNoOsc = '/Users/yuntse/data/coherent/NDSupernova/flux/grid//pinched_200289.dat'
fNH = '/Users/yuntse/data/coherent/NDSupernova/flux/grid/nh/pinched_200289.dat'
fIH = '/Users/yuntse/data/coherent/NDSupernova/flux/grid/ih/pinched_200289.dat'
```

```python
NoOsc = pd.read_csv( fNoOsc, header = None, sep = r'\s+', names = ['E', 'nue', 'numu', 'nutau', 'nuebar', 'numubar', 'nutaubar' ])
NH = pd.read_csv( fNH, header = None, sep = r'\s+', names = ['E', 'nue', 'numu', 'nutau', 'nuebar', 'numubar', 'nutaubar' ])
IH = pd.read_csv( fIH, header = None, sep = r'\s+', names = ['E', 'nue', 'numu', 'nutau', 'nuebar', 'numubar', 'nutaubar' ])
```

```python
fig, ax = plt.subplots( figsize = (8, 6) )
ax.plot( NoOsc['E']*1000, NoOsc['nue']*2/1e9, label = 'No Oscillation', lw = 2 )
ax.fill_between( NH['E']*1000, NH['nue']*2/1e9, 0, color = 'blue', alpha = 0.2, label = 'Normal Mass Ordering', lw = 0 )
ax.fill_between( IH['E']*1000, IH['nue']*2/1e9, 0, color = 'red', alpha = 0.2, label = 'Inverted Mass Ordering', lw = 0 )

ax.set_xlabel( 'Neutrino Energy [MeV]', fontsize = 20 )
ax.tick_params( axis = 'x', labelsize = 20 )
ax.set_ylabel( r'Fluence [$10^9$ $\nu$/0.2 MeV/cm$^2$]', fontsize = 20 )
ax.tick_params( axis = 'y', labelsize = 20 )
ax.yaxis.get_offset_text().set_fontsize( 20 )
ax.legend( loc = 'upper right', ncol = 1, fontsize = 20 )
# ax2.grid( which = 'both', linestyle = '--' )
ax.set_xlim([0, 50])
ax.set_ylim([0, 10.5])
fig.subplots_adjust( left = 0.15, right = 0.96, top = 0.93, bottom = 0.16)
fig.savefig('SNEnergy.pdf')
```
